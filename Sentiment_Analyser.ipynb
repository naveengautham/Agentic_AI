```python
from pathlib import Path
from autogen import ConversableAgent
from autogen.coding import LocalCommandLineCodeExecutor
```

```python
from Util import get_openai_api_key
OPEN_API_KEY=get_openai_api_key()
```

```python
llm_config={
    "config_list":{
        "model":"gpt-3.5-turbo",
        "api_key":OPEN_API_KEY
    }
}
```

```python
wor_dir= Path('sentiment')
wor_dir.mkdir(exist_ok=True)
executor= LocalCommandLineCodeExecutor(work_dir=wor_dir)
```

```python
executor= ConversableAgent(
    name="executor",
    llm_config=False,
    code_execution_config={
        "executor":executor
    },
    human_input_mode="NEVER"
)
```

```python
code_system_message="""
You are a sentiment analysis assistant. Your task is to analyze the sentiment of a list of text reviews using the VADER (Valence Aware Dictionary and sEntiment Reasoner) sentiment analysis tool from the `nltk.sentiment` library.

Input data:
[
    "I love the new feature, it's amazing!",
    "This update is terrible and frustrating.",
    "The service was okay, not the best.",
    "I’m so happy with the experience!",
    "Very disappointed. Will not recommend."
]

Instructions:
1. Use the VADER `SentimentIntensityAnalyzer` from `nltk.sentiment.vader`.
2. For each text review, calculate the sentiment scores: 
   - `pos` (positive)
   - `neu` (neutral)
   - `neg` (negative)
   - `compound` (overall sentiment)
3. Classify each review as **Positive**, **Negative**, or **Neutral** based on the compound score:
   - Positive: compound ≥ 0.05
   - Negative: compound ≤ -0.05
   - Neutral: otherwise
4. Display each review with its corresponding sentiment scores and final sentiment label.
5. Summarize the overall distribution of sentiment across all reviews in an tabular format.

Ensure the output is clear and easy to interpret.
"""
```

```python
writer= ConversableAgent(
    name="writer",
    system_message=code_system_message,
    llm_config=llm_config,
    code_execution_config=False,
    max_consecutive_auto_reply=2,
    human_input_mode="NEVER"
)
```

```python
import pprint
sentiment_results= executor.initiate_chat(writer,message="Do sentiment analysis by using VADAR and display the results in an tabular format for each reviews in an list")
pprint.pprint(sentiment_results)
```
