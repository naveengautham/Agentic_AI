```python
from Util import get_openai_api_key
OPEN_API_KEY=get_openai_api_key()
```

```python
llm_config={
    "config_list":{
        "model":"gpt-3.5-turbo",
        "api_key":OPEN_API_KEY
    }
}
```

```python
from autogen import UserProxyAgent,AssistantAgent,GroupChat,GroupChatManager

#agent1
user_proxy=UserProxyAgent(
    name="user_proxy",
    human_input_mode="NEVER",
    code_execution_config={
        "use_docker":False
    }
)
```

```python
headline_agent=AssistantAgent(
    name="headline_agent",
    llm_config=llm_config,
    system_message="""
    Fetch top 10 latest headlines from a given RSS feed URL or news API.
    Output as a list of dicts: [{"title": ..., "language": ..., "source": ...}]
    """
)
```

```python
translator_agent=AssistantAgent(
    name="translator_agent",
    llm_config=llm_config,
    system_message="""
    You are an translator agent.
    Translate the text into english if it is not in english.
    """
)
```

```python
analyze_sentiment=AssistantAgent(
    name="analyze_sentiment",
    llm_config=llm_config,
    system_message="""
    You are an analyze sentiment agent,
    analyze the headline of each topic i.e. positive,negative or neutral by using sentiment analysis and create one column "Sentiment analysis report" to each entry.
    And add the content of each topic below the Sentiment analysis report
    """
)
```

```python
group_chat=GroupChat(
    agents=[
        user_proxy,
        headline_agent,
        translator_agent,
        analyze_sentiment
    ],
    messages=[],
    max_round=8
)

manager=GroupChatManager(
    groupchat=group_chat,
    llm_config=llm_config
)
```

```python
response= user_proxy.initiate_chat(
    manager,
    message="Start headline processing from https://news.google.com/rss"
)
```
