```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Input data
documents = [
    "ChatGPT is an LLM model.",
    "ChatGPT is useful for NLP tasks.",
    "NLP involves working with language."
]

# Create the CountVectorizer object
vectorizer = CountVectorizer()

# Fit and transform the data
X = vectorizer.fit_transform(documents)

# Get the feature names (vocabulary terms)
feature_names = vectorizer.get_feature_names_out()

# Create a DataFrame for the count matrix
count_matrix = pd.DataFrame(X.toarray(), columns=feature_names)

# Output the shape of the matrix
print("Shape of the count matrix:", count_matrix.shape)
print("\nCount Matrix:")
print(count_matrix)

# Summarize the total frequency of each word
word_frequencies = count_matrix.sum()
print("\nWord Frequencies:")
print(word_frequencies)
```
